# Working with Missing Data

```python
#Checking for missing values using isnull() and notnull() 
```

```python
import pandas as pd
import numpy as np
 
dict = {'First Score':[100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score':[np.nan, 40, 80, 98]}
 
df = pd.DataFrame(dict)
```

```python
df
```

```python
df.isnull()  # Where is the nul , print True
```

```python
#Filling missing values using fillna(), replace() and interpolate()
```

```python
df.fillna(0) # Filling missing vlue with 0
```

```python
df.fillna(2) # Filling missing vlue with 2
```

```python
x=df['First Score'].mean()
df.fillna(x) # Filling missing vlue with average of First Score
```

```python
df.fillna(df['First Score'].mean()) # Filling missing vlue with average of First Score
```

```python
import pandas as pd

# Create a sample DataFrame
data = {'Column1': [1, 2, 3, None, 5],
        'Column2': [None, 6, 7, 8, 9]}
df = pd.DataFrame(data)

# Replace NaN values in 'Column1' with a specific value, e.g., 0
df['Column1'] = df['Column1'].fillna(0)

# Replace NaN values in 'Column2' with the mean of the column
mean_value = df['Column2'].mean()
df['Column2'] = df['Column2'].fillna(mean_value)

# Print the updated DataFrame
print(df)
```
